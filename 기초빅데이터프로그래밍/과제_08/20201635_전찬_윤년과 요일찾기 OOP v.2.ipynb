```python
#year, month, day input을 받는 함수이다.
def input_date():
    year = int(input("__년도를 입력하시오:"))
    month = int(input("__월을 입력하시오:"))
    day = int(input("__일을 입력하시오:"))
    
    return year, month, day


class DayName:
    #모든 instance에서 동일하게 사용할 수 있는 class 변수이다.
    day_per_month = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    date = ["월", "화", "수", "목", "금", "토", "일"]
    
    def __init__(self, year, month, day):
        self._year = year
        self._month = month
        self._day = day
    
    #property decorator을 통해 year, month, day 를 보호
    @property
    def year(self):
        return self._year
    
    @property
    def month(self):
        return self._month
    
    @property
    def day(self):
        return self._day
    
    # 윤년 계산 method
    def is_leap(self, *args):
        # is_leap method에 input이 없는 경우, self.year을 바탕으로 계산한다.
            if len(args) == 0:
                
                #400으로 나누어지는 해는 윤년이다.
                if self.year%400 == 0:
                    return True
    
                #100으로 나누어지는 해는 윤년이 아니다.
                elif self.year%100 == 0:
                    return False
    
                #4로 나누어지는 해는 윤년이다.
                elif self.year%4 == 0:
                    return True
    
                else:
                    return False    
            
            # is_leap method에 input이 존재하는 경우, args[0]을 바탕으로 계산한다.
            else:
                
                #400으로 나누어지는 해는 윤년이다.
                if args[0]%400 == 0:
                    return True
    
                #100으로 나누어지는 해는 윤년이 아니다.
                elif args[0]%100 == 0:
                    return False
    
                #4로 나누어지는 해는 윤년이다.
                elif args[0]%4 == 0:
                    return True
    
                else:
                    return False  

    def get_name(self):
         #2월 29일이며 윤년이 아닌 경우, return "입력하신 날짜는 존재하지 않습니다."를 수행한다.
         #이외의 경우, return "_요일"
         #추가로 윤년인 경우 "입력하신 %s은 윤년입니다" %year을 수행한다. 
    
        #총 일수를 세는 변수이다.
        day_count = 0
    
        if self.is_leap() == False:
            if self.month==2 and self.day == 29:
                return "입력하신 날짜는 존재하지 않습니다."
    
        #요일을 계산한다.
        #year에 의한 일수를 계산한다.
        for i in range(1, self.year):
            if self.is_leap(i) == True:
                day_count += 366
            else:
                day_count += 365
    
        #month에 의한 일수를 계산한다.
        for i in range(1, self.month):
            day_count += DayName.day_per_month[i]
            if i == 2 and self.is_leap() == True:
                day_count += 1
    
        #day에 의한 일수를 계산한다.
        day_count += (self.day - 1) #1일부터 시작하므로, 1을 빼준다.
        
    
        #총 day_count를 7로 나누어서 date를 통해 무슨 요일인지 파악한다.
        
        if self.is_leap() == True:
            return DayName.date[day_count%7] + "요일\n" + "입력하신 %s은 윤년입니다." %year
        else:
            return DayName.date[day_count%7] + "요일"


    @staticmethod
    def leap_count(date1, date2):
        count = 0
        for i in range(date1.year, date2.year+1):
            if date1.is_leap(i):
                count += 1
        
        print("there are", count, "leap_years between", date1.year, "and", date2.year)
    
    @staticmethod
    def day_count(date1, date2):
        
        date1_count = 0
        if date1.is_leap() == False:
            if date1.month==2 and date1.day == 29:
                print("첫 번째 DayName instance의 날짜는 존재하지 않습니다.")
                return

        for i in range(1, date1.year):
            if date1.is_leap(i) == True:
                date1_count += 366
            else:
                date1_count += 365
    
        #month에 의한 일수를 계산한다.
        for i in range(1, date1.month):
            date1_count += DayName.day_per_month[i]
            if i == 2 and date1.is_leap() == True:
                date1_count += 1
    
        #day에 의한 일수를 계산한다.
        date1_count += (date1.day - 1)
        
        date2_count = 0
        if date2.is_leap() == False:
            if date2.month==2 and date2.day == 29:
                print("두 번째 DayName instance의 날짜는 존재하지 않습니다.")
                return

        for i in range(1, date2.year):
            if date2.is_leap(i) == True:
                date2_count += 366
            else:
                date2_count += 365
    
        #month에 의한 일수를 계산한다.
        for i in range(1, date2.month):
            date2_count += DayName.day_per_month[i]
            if i == 2 and date2.is_leap() == True:
                date2_count += 1
    
        #day에 의한 일수를 계산한다.
        date2_count += (date2.day - 1)
        
        print("두 날짜 사이에", date2_count-date1_count, "일이 있습니다.")
    
    
    # 해당 년도의 13일의 금요일을 알아내는 class method
    @classmethod
    def friday_13_count(cls, input_year):
        count_13 = 0
        day_count = 0
        
        for i in range(1, input_year):
            if i%400==0 or (i%4==0 and i%100!=0):
                day_count += 366
            else:
                day_count += 365
        
        for i in range(0, 12):
            day_count += cls.day_per_month[i]
            if i == 2 and (input_year%400==0 or (input_year%4==0 and input_year%100!=0)):
                day_count += 1
            
            if day_count%7 == 6:
                count_13 += 1
        
        print("해당 년에 13일의 금요일이", count_13, "개 존재합니다.")
    
#DayName을 상속받는 Person class, year, month, day 는 태어난 날을 의미한다.
class Person(DayName):
    def __init__(self, name, year, month, day):
        super().__init__(year, month, day)
        self._name = name
        
    @property
    def name(self):
        return self._name
    
    def get_age(self):
        import time
        today = time.localtime(time.time())
        return len(range(self.year, today.tm_year +1))
    
    def profile(self):
        print("이름:", self.name, "나이:", self.get_age())
        print("생년월일:", self.year, self.month, self.day)
```

```python
if __name__ =="__main__":
    year, month, day = input_date()
    day_name = DayName(year, month, day)
    print(day_name.get_name()) 
```

```python
day1 = DayName(1981, 12, 1)
day2 = DayName(2022, 3, 14)

# 구현한 staticmethod(윤년, 날짜 count)의 적용
DayName.leap_count(day1, day2)
DayName.day_count(day1, day2)

# 구현한 classmethod(13일의 금요일 count)의 적용
DayName.friday_13_count(2022)
```

```python
# DayName을 상속받는 Person class 의 적용
person = Person("전찬", 2001, 6, 27)
print(person.get_age())
person.profile()
```
